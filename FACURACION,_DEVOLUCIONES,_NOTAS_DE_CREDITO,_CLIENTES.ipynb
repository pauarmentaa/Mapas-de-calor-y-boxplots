## **HISTOGRAMA 1**
>El objetivo de este histograma es visualizar la distribución de la cantidad total de devoluciones por año. Cada barra en el histograma representa un rango de valores de cantidad total, y la altura de la barra indica la frecuencia de ocurrencia de esos valores en el conjunto de datos.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Agregar una columna "Año" que represente el año de la devolución
dev["Año"] = pd.to_datetime(dev["FECHA_DOC"]).dt.year
# Extraer la columna "CAN_TOT"
can_tot = dev["CAN_TOT"]
# Crear el histograma
plt.hist(can_tot, bins=10, range=(0, 200000), edgecolor='black', alpha=0.5, label="Total")
# Obtener los años únicos presentes en el DataFrame
años = dev["Año"].unique()
# Crear un histograma por cada año
for año in años:
    can_tot_año = dev[dev["Año"] == año]["CAN_TOT"]
    plt.hist(can_tot_año, bins=10, range=(0, 200000), edgecolor='black', alpha=0.5, label=str(año))
# Personalizar el gráfico
plt.title("Histograma de Cantidad Total por Año")
plt.xlabel("Cantidad Total")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

## **HISTOGRAMAS**
>El objetivo de este código es crear un conjunto de histogramas, uno por cada año presente en el DataFrame, para visualizar la distribución de la cantidad total de devoluciones en cada año de manera comparativa.

>Al utilizar subplots, se crea un conjunto de subplots apilados verticalmente, donde cada subplot representa un año específico. El número de subplots y su tamaño se ajustan dinámicamente según la cantidad de años presentes en el DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Agregar una columna "Año" que represente el año de la devolución
dev["Año"] = pd.to_datetime(dev["FECHA_DOC"]).dt.year
# Extraer la columna "CAN_TOT"
can_tot = dev["CAN_TOT"]
# Obtener los años únicos presentes en el DataFrame
años = dev["Año"].unique()
# Crear un subplot por cada año
fig, ax = plt.subplots(len(años), 1, figsize=(8, len(años) * 4), sharex=True)
# Iterar sobre cada año y crear el histograma correspondiente
for i, año in enumerate(años):
    can_tot_año = dev[dev["Año"] == año]["CAN_TOT"]
    ax[i].hist(can_tot_año, bins=10, edgecolor='black', alpha=0.5, label=str(año))
    ax[i].set_title("Histograma de Cantidad Total para el año {}".format(año))
    ax[i].set_xlabel("Cantidad Total")
    ax[i].set_ylabel("Frecuencia")
    ax[i].set_xlim(0, 200000)
    ax[i].legend()
# Ajustar los espacios entre los subplots
plt.tight_layout()
plt.show()
```

## **MAPA DE CALOR**
>El objetivo de este mapa de calor es visualizar la matriz de correlación entre las variables "CAN_TOT" (monto total) y "DES_TOT" (descuento total) de las devoluciones.

>La matriz de correlación muestra la relación estadística entre pares de variables. En este caso, el mapa de calor representa los coeficientes de correlación entre las variables seleccionadas. Los coeficientes de correlación varían entre -1 y 1, donde 1 indica una correlación positiva perfecta, -1 indica una correlación negativa perfecta y 0 indica la ausencia de correlación.

>El mapa de calor utiliza colores para resaltar la fuerza y la dirección de la correlación entre las variables. En este caso, se utiliza la paleta de colores 'coolwarm', donde los tonos azules indican una correlación negativa, los tonos rojos indican una correlación positiva y los tonos más claros indican una correlación cercana a cero.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Selecciona las variables relevantes para el cálculo de la correlación
variables = ['CAN_TOT', 'DES_TOT']
dev_selected = dev[variables]
# Calcula la matriz de correlación
correlation_matrix = dev_selected.corr()
# Crea un mapa de calor utilizando seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de correlación')
plt.show()
```

## **Diagrama de cajas y bigotes (Boxplot)**
>El objetivo principal del boxplot es visualizar la distribución de una variable numérica y resumir sus características principales de manera concisa. Al crear un boxplot, se busca obtener información sobre la simetría de la distribución, la presencia de valores atípicos y la concentración de los datos alrededor de ciertos valores.

Los elementos del boxplot incluyen:

La línea horizontal en el medio de la caja representa la mediana.
La caja muestra el rango intercuartil (IQR), que abarca desde el primer cuartil (Q1) hasta el tercer cuartil (Q3).
Los bigotes muestran el rango de los datos dentro de 1.5 veces el IQR.
Los puntos fuera de los bigotes se consideran valores atípicos o anomalías.

```python
#Obtenemos los valores de la columna "CAN_TOT" y guardamos
montos_totales = ncred['CAN_TOT'].tolist()
#Creamos el boxplot
plt.boxplot(montos_totales)
plt.xlabel('Variable CAN_TOT')
plt.ylabel('Montos Totales')
plt.title('Boxplot de Montos Totales de Notas de Crédito')
plt.show()
```

## **Histogramas**
>El objetivo de los dos histogramas que realizamos es analizar la distribución de los status de las notas de crédito en el primer histograma y la distribución de los estados de las notas de crédito por año en el segundo histograma. Estas visualizaciones nos permiten obtener información sobre la frecuencia de cada status y cómo varía a lo largo del tiempo.

>En el primer histograma, al mostrar la frecuencia de cada status en las notas de crédito, podemos identificar cuáles son los estados más comunes y cuáles son menos frecuentes. Esto puede proporcionar información útil sobre la eficiencia del proceso de aprobación o el nivel de incidencias en las notas de crédito.

```python
#Calcula la frecuencia de cada estado en la columna STATUS
frecuencias = ncred['STATUS'].value_counts().reset_index()
frecuencias.columns = ['STATUS', 'Frecuencia']
frecuencias = frecuencias.sort_values('STATUS')
#Crear el histograma
plt.bar(frecuencias['STATUS'], frecuencias['Frecuencia'])
plt.xlabel('STATUS')
plt.ylabel('Frecuencia')
plt.title('Histograma de STATUS de notas de crédito')
plt.show()
```


>En el segundo histograma, al agregar la dimensión temporal y mostrar la frecuencia de los status por año, podemos observar cómo ha evolucionado la distribución de los status a lo largo del tiempo. Esto nos puede ayudar a identificar tendencias o patrones en los status de las notas de crédito a lo largo de los años, lo cual puede ser útil para tomar decisiones o identificar posibles mejoras en los procesos internos de la empresa.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Agregar una columna "Año" que represente el año de la devolución
ncred['FECHA_DOC'] = pd.to_datetime(ncred['FECHA_DOC'])
ncred['AÑO'] = ncred['FECHA_DOC'].dt.year
#Calcular la frecuencia de cada estado por año utilizando groupby en el DataFrame ncred
frecuencias_por_año = ncred.groupby(['AÑO', 'STATUS']).size().reset_index(name='Frecuencia')
# Crear un histograma para cada estado por año
for estado in ncred['STATUS'].unique():
    datos_estado = frecuencias_por_año[frecuencias_por_año['STATUS'] == estado]
    plt.bar(datos_estado['AÑO'], datos_estado['Frecuencia'], label=estado)

plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.title('Histograma de estado de notas de crédito por año')
plt.legend()
plt.show()
```
